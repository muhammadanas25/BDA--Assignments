```
from pyspark.sql import SparkSession 

spark=SparkSession.builder.master('local[*]').appName('my-app').getOrCreate()
spark
```

```
data=spark.read.format('csv').option('inferschema','true').option('header','true').load('/FileStore/tables/employee_earnings_report_2016.csv')
```

```
data.show()
```

```
import pyspark.sql.functions as f
#first removing $ and , signs from total_earning column
data=data.select(("*"), f.translate(f.col("TOTAL EARNINGS"), "$,", "").alias("Earning"))
data.printSchema()
```

```
# changing earning column to numeric
data2 = data.withColumn('Earnings',
								data['Earning']
								.cast('integer'))
data2.printSchema()
```

```
from pyspark.sql.functions import col, asc,desc,avg
Highest_avg_dept=data2.groupby('DEPARTMENT_NAME').avg('Earnings').orderBy(col('avg(Earnings)'),ascending=False).limit(1).show()
```

```
data.select('Name').where(data.DEPARTMENT_NAME=='Asst Superintendent-Network A').show()
```

```
#task 2Find top 3 departments in which employees collectively have earned more money through
#overtime.
```

```
#converting overtime column to float type by first removing $and , signs
import pyspark.sql.functions as f
data=data.select(("*"), f.translate(f.col("OVERTIME"), "$,", "").alias("overtime_earn"))
data1 = data.withColumn('OVERTIME',
								data['overtime_earn']
								.cast('float'))
```

```

data1.printSchema()
#overtime changed to float
```

```
data1.groupby('Department_name').sum('OVERTIME').orderBy(col('sum(OVERTIME)'),ascending=False).limit(3).show()
```

```
data1.groupby(['Department_name','GENDER']).sum('OVERTIME').orderBy(col('sum(OVERTIME)'),ascending=False).limit(3).show()
```

```
data2.createOrReplaceTempView("people")
sqlDF = spark.sql('''select name from people where department_name=(select department_name from (SELECT DEPARTMENT_NAME ,avg(earnings) as avg FROM people group by DEPARTMENT_NAME order by avg desc limit 1))

''')
sqlDF.show(truncate=0)
```

```
#2
data1.createOrReplaceTempView("people")
sqldf2=spark.sql(''' 
select department_name,sum(overtime) as summ from people group by department_name order by summ desc limit 3

''')
sqldf2.show(truncate=0)
```

```

#3sql
sqldf3=spark.sql(''' 
select department_name,gender,sum(overtime) as summ from people group by department_name,gender order by summ desc limit 3

''')
sqldf3.show(truncate=0)
```
